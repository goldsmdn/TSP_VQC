```python
import numpy as np

# Import Qiskit
from qiskit import QuantumCircuit, transpile

circ = QuantumCircuit(2)
circ.h(0)
circ.cx(0, 1)
circ.measure_all()

# Transpile for simulator
simulator = AerSimulator(method='statevector',device='GPU')
print(simulator.available_devices())
circ = transpile(circ, simulator)

# Run and get counts
result = simulator.run(circ).result()
counts = result.get_counts(circ)
```

```python
from qiskit_ibm_runtime import QiskitRuntimeService

print("=== Checking Aer Simulator Devices ===")
try:
    sim = AerSimulator()
    devices = sim.available_devices()
    print("Available devices:", devices)
    if "GPU" in str(devices):
        print("✅ GPU support detected in Aer!")
    else:
        print("⚠️ Only CPU detected — GPU backend not active.")
except Exception as e:
    print("❌ AerSimulator check failed:", e)

print("\n=== Checking IBM Runtime Service ===")
try:
    service = QiskitRuntimeService()  # Assumes you’ve saved your IBM Quantum API token
    backends = service.backends()
    print(f"✅ IBM Runtime loaded successfully. {len(backends)} backends available.")
except Exception as e:
    print("⚠️ Could not access IBM Runtime:", e)
    print("   (If this is your first time, run: QiskitRuntimeService.save_account('<YOUR_IBM_API_TOKEN>'))")
```
