```python
def find_sequence(bit_string, locs, gray):
    f = math.factorial(locs)

    bit_string_length = math.ceil(math.log2(f))
    if len(bit_string) != bit_string_length:
        print('incorrect_bit_string_length')
        print(bit_string_length)

    f = math.factorial(locs)
    x = convert_binary_list_to_integer(bit_string, gray=gray)
    #print(f'x = {x}')

    y = x % f
    end_cycle_list = []
    start_cycle_list = [i for i in range(locs)]
    i = 0
    #print(end_cycle_list, start_cycle_list)

    #if x >= f:
    #    #algorithm won't work
    #    end_cycle_list = start_cycle_list
    #else:
    #print(f'y = {y}')  
    while i < locs:
        f = f / (locs - i)
        #print(f'f = {f}')
        #correcting mistype in paper
        #k = math.floor(x / f)
        k = math.floor(y / f)
        #print(f'k = {k}')
        #print(end_cycle_list, start_cycle_list)
        end_cycle_list.append(start_cycle_list[k])
        start_cycle_list.remove(start_cycle_list[k])
        #print(end_cycle_list, start_cycle_list)
        #correcting mistype in paper
        #x -= k * f
        y -= k * f
        i += 1

    return end_cycle_list
```
